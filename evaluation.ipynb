```python
import os
from k_fit_all import approximate, k1, k2, open_file
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.listdir('/home/quantum/Documents/iltpe/')
```

```python
def k_approx(path):
    data = open_file(path)
    T = data['T'].astype('float')
    c1, c2, n, c3, c4, k_max = approximate(data)
    print('C1 {}, C2 {}, N {}, C3 {}, C4 {}'.format(c1, c2, n, c3, c4, T))
    print(k_max)
    k_fit = k1(T, c1, c2, n) + k2(T, c3, c4)

    plt.scatter(data['T'], data['k'], color='g')
    plt.plot(T, k_fit*k_max)
    plt.yscale('log')
    plt.xscale('log')
```

```python
k_approx('/home/quantum/Documents/iltpe/Cu10Zn2Sb4S13.dat')
```
